##**Proyecto Desafío 1: Ventas Online**

El objetivo de este proyecto es analizar los resultados de un evento con los clientes de una empresa de venta online. Se recopiló un conjunto de datos que contiene los clientes que más gastaron en productos durante los 5 días de ventas, que es la duración del evento. Este análisis identificará al cliente con la mayor compra esta semana, quien recibirá un premio de la tienda, y posteriormente, puede ayudar a la empresa a crear nuevas estrategias para atraer más clientes.

```python
import pandas as pd
```

```python
Archivo_ventas = '/content/drive/MyDrive/raw data Transformacion_manipulacion/dados_vendas_clientes.json'

datos_ventas = pd.read_json(Archivo_ventas)
datos_ventas
```

```python
Datos_ventas_normalize = pd.json_normalize(datos_ventas['dados_vendas'])
Datos_ventas_normalize
```

###**Eliminar datos en listas dentro del DataFrame**

```python
columns_ventas = list(Datos_ventas_normalize.columns)
columns_ventas
```

```python
Datos_ventas_expande = Datos_ventas_normalize.explode(columns_ventas[1:])
Datos_ventas_expande.head()
```

```python
Datos_ventas_expande.reset_index(inplace=True,drop=True)
Datos_ventas_expande.head()
```

###**Verificar tipos de datos**

```python
Datos_ventas_expande.info()
```

```python
import numpy as np
Datos_ventas_expande['Valor da compra'] = Datos_ventas_expande['Valor da compra'].apply(lambda x: x.replace('R','').replace(',','.').strip())
Datos_ventas_expande.sample(5)
```

```python
Datos_ventas_expande['Valor da compra'] = Datos_ventas_expande['Valor da compra'].astype(np.float64)
```

```python
Datos_ventas_expande.info()
```

###**Datos textuales**

```python
Datos_ventas_expande['Cliente'] = Datos_ventas_expande['Cliente'].str.lower()
Datos_ventas_expande.head()
```

```python
Datos_ventas_expande['Cliente'] = Datos_ventas_expande['Cliente'].str.replace('[^a-z]',' ',regex=True)
Datos_ventas_expande.head()
```

```python
Datos_ventas_expande['Cliente'] = Datos_ventas_expande['Cliente'].str.replace('(?<!\w)-(?!\w)','',regex=True)
Datos_ventas_expande.head()
```

```python
Datos_ventas_expande['Cliente'] = Datos_ventas_expande['Cliente'].str.split()
Datos_ventas_expande.head()
```
