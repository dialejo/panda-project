##**Proyecto Desafío 1: Ventas Online**

El objetivo de este proyecto es analizar los resultados de un evento con los clientes de una empresa de venta online. Se recopiló un conjunto de datos que contiene los clientes que más gastaron en productos durante los 5 días de ventas, que es la duración del evento. Este análisis identificará al cliente con la mayor compra esta semana, quien recibirá un premio de la tienda, y posteriormente, puede ayudar a la empresa a crear nuevas estrategias para atraer más clientes.

```python
import pandas as pd
```

```python
Archivo_ventas = '/content/drive/MyDrive/raw data Transformacion_manipulacion/dados_vendas_clientes.json'

datos_ventas = pd.read_json(Archivo_ventas)
datos_ventas
```

```python
Datos_ventas_normalize = pd.json_normalize(datos_ventas['dados_vendas'])
Datos_ventas_normalize
```

###**Eliminar datos en listas dentro del DataFrame**

```python
columns_ventas = list(Datos_ventas_normalize.columns)
columns_ventas
```

```python
Datos_ventas_expande = Datos_ventas_normalize.explode(columns_ventas[1:])
Datos_ventas_expande.head()
```

```python
Datos_ventas_expande.reset_index(inplace=True,drop=True)
Datos_ventas_expande.head()
```

###**Verificar tipos de datos**

```python
Datos_ventas_expande.info()
```

```python
import numpy as np
Datos_ventas_expande['Valor da compra'] = Datos_ventas_expande['Valor da compra'].apply(lambda x: x.replace('R','').replace(',','.').strip())
Datos_ventas_expande.sample(5)
```

```python
Datos_ventas_expande['Valor da compra'] = Datos_ventas_expande['Valor da compra'].astype(np.float64)
```

```python
Datos_ventas_expande.info()
```

###**Datos textuales**

```python
Datos_ventas_expande['Cliente'] = Datos_ventas_expande['Cliente'].str.lower()
Datos_ventas_expande.head()
```

```python
Datos_ventas_expande['Cliente'] = Datos_ventas_expande['Cliente'].str.replace('[^a-z]',' ',regex=True)
Datos_ventas_expande.head()
```

```python
Datos_ventas_expande['Cliente'] = Datos_ventas_expande['Cliente'].str.replace('(?<!\w)-(?!\w)','',regex=True)
Datos_ventas_expande.head()
```

```python
Datos_ventas_expande.head()
```

###Datos fecha

El objetivo de este proyecto es analizar los resultados de un evento con los clientes de una empresa de venta online. Se recopiló un conjunto de datos que contiene los clientes que gastaron más en productos dentro de los 5 días posteriores a la venta, que es la duración del evento. Este análisis identificará al cliente con la mayor compra esa semana, quien recibirá un premio de la tienda, y posteriormente, puede ayudar a la empresa a crear nuevas estrategias para atraer más clientes.


```python
import datetime

Datos_ventas_expande['Data de venda'] = pd.to_datetime(Datos_ventas_expande['Data de venda'])
Datos_ventas_expande
```

```python
Subset_ventas = Datos_ventas_expande.groupby(Datos_ventas_expande['Cliente'])['Valor da compra'].sum()
Subset_ventas
```

##**2.Proyecto Desafío 2: Administración de Condominios**

Administrar condominios es una tarea que requiere mucha atención y organización. Entre las diversas responsabilidades de gestión se encuentra el cobro del alquiler a los inquilinos. Para garantizar la buena salud financiera de la empresa, es fundamental que estos pagos se realicen de forma regular y puntual. Sin embargo, sabemos que esto no siempre sucede.

```python
Archivo_condominios = '/content/drive/MyDrive/raw data Transformacion_manipulacion/dados_locacao_imoveis.json'

Datos_condominios = pd.read_json(Archivo_condominios)
Datos_condominios
```

```python
Datos_condominios_normalize = pd.json_normalize(Datos_condominios['dados_locacao'])
Datos_condominios_normalize
```

```python
columns_condominios = list(Datos_condominios_normalize.columns)
columns_condominios
```

```python
Datos_condominios_expande = Datos_condominios_normalize.explode(columns_condominios[1:])
Datos_condominios_expande.head()
```

```python
Datos_condominios_expande.reset_index(inplace=True,drop=True)
Datos_condominios_expande.head()
```

```python
Datos_condominios_expande.info()
```

```python
Datos_condominios_expande['valor_aluguel'] = Datos_condominios_expande['valor_aluguel'].apply(lambda x: x.replace(' reais','').replace('$ ','').replace(',','').strip())
Datos_condominios_expande.sample(5)
```

```python
Datos_condominios_expande['valor_aluguel'] = Datos_condominios_expande['valor_aluguel'].astype(np.float64)
Datos_condominios_expande.info()
```

```python
Datos_condominios_expande['apartamento'] = Datos_condominios_expande['apartamento'].str.replace('()','')
Datos_condominios_expande.head()
```

### Datos texto

Administrar condominios es una tarea que requiere mucha atención y organización. Entre las diversas responsabilidades de gestión se encuentra el cobro del alquiler a los inquilinos. Para garantizar la buena salud financiera de la empresa, es fundamental que estos pagos se realicen de forma regular y puntual. Sin embargo, sabemos que esto no siempre sucede. Teniendo esto en cuenta, propongo un desafío de procesamiento de datos con el objetivo de analizar el retraso en el pago de la renta en el condominio ficticio de algunos residentes.


```python
import datetime
Datos_condominios_expande['datas_combinadas_pagamento'] = pd.to_datetime(Datos_condominios_expande['datas_combinadas_pagamento'], format='%d/%m/%Y')
Datos_condominios_expande['datas_de_pagamento'] = pd.to_datetime(Datos_condominios_expande['datas_de_pagamento'],format='%d/%m/%Y')
```

```python
Datos_condominios_expande.info()
```

```python
Datos_condominios_expande['atrasos'] = (Datos_condominios_expande['datas_de_pagamento'] - Datos_condominios_expande['datas_combinadas_pagamento']).dt.days
Datos_condominios_expande.head()
```

```python
media_atrasos = Datos_condominios_expande.groupby(['apartamento'])['atrasos'].mean()
media_atrasos
```
