Hay varias formas de crear columnas en un DataFrame en Pandas. Algunos ejemplos son:

1. Asignacin directa de valores a una nueva columna:

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = [7, 8, 9]
df
```

2. Creacin de una columna a partir de operaciones entre otras columnas:

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'] + df['B']

df
```

3. Uso del mtodo assign() para crear una nueva columna:

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df = df.assign(C=[7, 8, 9])
df
```

4. Uso del mtodo apply() para aplicar una funcin a una columna existente y crear una nueva columna:

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'].apply(lambda x: x * 2)
df
```

1. Los estudiantes participaron en una actividad extracurricular y ganaron puntos extras. Estos puntos extras corresponden al 40% de su nota actual. Por lo tanto, crea una columna llamada "Puntos_extras" que contenga los puntos extras de cada estudiante, es decir, el 40% de su nota actual.

```python
df1
```

```python
datos1['Puntos_extras'] = datos1['Nota'].apply(lambda x: x * 0.4)
datos1
```

2. Crea otra columna llamada "Notas_finales" que contenga las notas de cada estudiante sumadas con los puntos extras.

```python
datos1['Notas_finales'] = datos1['Nota'] + datos1['Puntos_extras']
datos1
```

3. Dado que hubo una puntuacin extra, algunos estudiantes que no haban sido aprobados antes pueden haber sido aprobados ahora. En funcin de esto, crea una columna llamada "Aprobado_final" con los siguientes valores:


True: si el estudiante est aprobado (la nota final debe ser mayor o igual a 7.0).

False: si el estudiante est reprobado (la nota final debe ser menor que 7.0).

```python
datos1['Aprobado_final'] = datos1['Notas_finales'].apply(lambda x: 'True' if x >= 7 else 'False')
datos1
```
