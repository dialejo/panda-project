#**Visualización de datos con python: Matplotlib, Seaborn y Plotly**

##**Importando los datos**

```python
import pandas as pd
Archivo_datos = '/content/drive/MyDrive/raw data Vizualizacion /inmigrantes_canada.csv'
df = pd.read_csv(Archivo_datos)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais',inplace=True)
df.tail(5)
```

Crear a lista desde 1980 hasta el 2013, map-función convierte cada numero en una cadena con el str

```python
anos = list(map(str,range(1980,2014)))
anos
```

df.loc busca en el dataframe el indice colombia y trae la info

```python
colombia = df.loc['Colombia',anos]
colombia
```

crear diccionario. año con index y inmigrantes con valores

```python
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes' : colombia.values.tolist()}
col_dict
```

```python
datos_colombia = pd.DataFrame(col_dict)
datos_colombia.head()
```

```python
datos_comparación = df.loc[['Brasil','Argentina'],anos]
```

```python
df_comparacion = datos_comparación.T
df_comparacion.head()
```

##**Graficando**

##**sugerencias visualizacion graficos**

1. Usar títulos claros y concisos: el título de cada subplot debe ser corto y descriptivo para que el lector pueda rápidamente entender lo que se está mostrando en cada elemento. Además, si estás comparando diferentes conjuntos de datos, puede ser útil añadir un subtítulo explicando lo que se está comparando.

2. Mantener la misma escala en los ejes: para evitar distorsiones en la comparación entre los subplots, es importante mantener la misma escala en los ejes X e Y en todos ellos. Esto se puede hacer utilizando las funciones set_xlim() y set_ylim().

3. Evitar la superposición de gráficos: es importante asegurarse de que cada subplot esté claramente separado de los demás, sin superposición. Si hay una superposición, la figura puede volverse confusa y difícil de entender. Una manera de evitar superposiciones es ajustar el tamaño de los subplots para que haya espacio suficiente entre ellos. Además, podemos añadir la función fig.subplots_adjust() que recibe el parámetro hspace. Controla el espaciado vertical entre los subplots y el parámetro wspace, que controla el espaciado horizontal. El valor por defecto de estos parámetros es 0.2, pero puedes ajustarlos según tus necesidades.

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
df_ventas = pd.DataFrame(ventas_2022, index=tiendas)
df_ventas
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,8))
plt.subplots_adjust(wspace=0.3,hspace=0.5)
fig.suptitle('Ventas mensuales por tienda desde Enero hasta Diciembre del 2022\nen las tiendas A, B, C y D',va='top',)

axs[0,0].plot(df_ventas.loc['A'])
axs[0,0].set_title(' Ventas en la tienda A')

axs[0,1].plot(df_ventas.loc['B'])
axs[0,1].set_title('Ventas en la tienda B')

axs[1,0].plot(df_ventas.loc['C'])
axs[1,0].set_title('Ventas en la tienda C')

axs[1,1].plot(df_ventas.loc['D'])
axs[1,1].set_title('Ventas en la tienda D')

ymin = 0
ymax = 500

for ax in axs.flat:
  ax.set_ylim(ymin,ymax)
  ax.xaxis.set_major_locator(plt.MultipleLocator(2))
  ax.set_xlabel('Meses')
  ax.set_ylabel('Ventas')
  ax.grid()

plt.show()
```

```python
sudamerica = df.query('Region == "América del Sur"')
sudamerica
```

```python
Colores = ['royalblue','Orange','forestgreen',
           'orchid','purple', 'brown','slateblue','gray',
           'navy','teal','tomato']

fig, ax = plt.subplots(figsize = (12,5))
ax.barh(sudamerica.index, sudamerica['Total'],color = Colores)
ax.set_title('inmigracion de Sudamericanos hacia Canada \ndesde 1980 hasta 2013',loc='left',fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Numero de Inmigrantes',fontsize=14)

ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values('Total',ascending=True)
sudamerica_sorted


fig, ax = plt.subplots(figsize = (12,5))
ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'],color = Colores)
ax.set_title('inmigracion de Sudamericanos hacia Canada \ndesde 1980 hasta 2013',loc='left',fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Numero de Inmigrantes',fontsize=14)

ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
Colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    Colores.append('ForestGreen')
  else:
    Colores.append('Silver')

fig, ax = plt.subplots(figsize = (12,5))
ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis='y',which = 'both', length = 0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'],color = Colores)
ax.set_title('Sudamérica: Colombia Segundo pais con mas migrantes \nhacia Canada desde 1980 hasta 2013',loc='left',fontsize=20)
ax.set_ylabel('')
#ax.set_xlabel('Numero de Inmigrantes',fontsize=14)

#ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i,str(j),color='black',fontsize=12,ha='left',va='center')
fig.savefig('/content/drive/MyDrive/Graphs/inmigrates_Sudamericanos_canada.png',transparent=False,dpi=300,bbox_inches='tight')
plt.show()
```
